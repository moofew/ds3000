```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression

df = pd.read_csv("DP_LIVE_26102023160103370.csv")
```

## Question 2: 

(10 pts) Extract the meat consumption data for any three countries of your choice and evaluate the patterns and trends in their meat consumption. You can explore any (2) types of meat (from the chosen countries), and you should only select one unit of measurement**. Ensure that you visualize the data and explain your findings.

```python
countries = ['AUS', 'CAN', 'JPN'] # the three countries of my choice
meats = ['BEEF', 'PIG'] # the two types of meat
unit = 'KG_CAP' # the unit of measurement

# filtered data frame to sort by the countries, meats, and unit
filtered_df = df[(df['LOCATION'].isin(countries)) & (df['SUBJECT'].isin(meats)) & (df['MEASURE'] == unit)]

# for each country and meat
for country in countries:
    for meat in meats:
        # create a df for the specific country + meat
        country_meat_df = filtered_df[(filtered_df['LOCATION'] == country) & (filtered_df['SUBJECT'] == meat)]
        # plot the df by year and weight
        plt.plot(country_meat_df['TIME'], country_meat_df['Value'], label=meat)

    plt.xlabel('Year') # label x axis
    plt.ylabel('Meat Consumption (kilograms/capita)') # label y axis
    plt.title(f'Meat Consumption Trends in {country}') # label chart
    plt.legend() # legend for readability
    plt.grid(True) # show grid for better readability
    plt.show() # show graph
```

The graphs above show the meat consumption trends of three countries (AUS, CAN, JPN). All three graphs have different trends. 

1) The meat consumption trends for AUS show that beef started at almost 30 kg/cap and has slowly decreased and is still decreasing. On the other hand, pig started at about 15 kg/cap and has been slowly rising and is now above beef.

2) The meat consumption trends for CAN is much different than AUS. In CAN, the trends between beef and pig are almost parallel to each other. By this, I mean that they almost look like parallel lines, as they both seem to decrease at almost the same rate since 1990. In this case, beef has always been consumed more and has been decreasing, whereas pig has been consumed slightly less and has also been decreasing.

3) The meat consumption trends for JPN are also sort of parallel. However, they are constant parallel lines, meaning they don't really decrease or increase. Pig consumption started at about 13 kg/cap and has risen to 16 kg/cap. On the other hand, beef started at 6 kg/cap and is now slightly below 8 kg/cap.

A finding that I found throughout all the graphs is that AUS and CAN have very similar consumption between pig and beef, whereas JPN has a big gap between the two meats. 

## Question 3: 

(5 pts) Extract the poultry consumption data, from 1994 to 2014 (inclusive), for Mexico, that is measured in thousand tonnes of carcass weight**. 

Note: the filtered dataframe should only contain the time and the values.

```python
country = 'MEX' # set country to Mexico
meat = 'POULTRY' # set meat type to poultry
unit = 'THND_TONNE' # set unit to thousand tonnes 
start_year, end_year = 1994, 2014 # set start and end years

# filter the data frame by country, meat, unit, start year + end year. 
mex_poultry_df = df[(df['LOCATION'] == country) & (df['SUBJECT'] == meat) & (df['MEASURE'] == unit) & (df['TIME'] >= start_year) & (df['TIME'] <= end_year)][['TIME', 'Value']]

mex_poultry_df.head(21)
```

## Question 6: 

(15 pts) Predict poultry consumption from 2015 to 2020 using a simple linear regression model.

1. Ensure that you visualize the original dataset along with the model's predictions.
2. Evaluate the model's predictions against the actual consumption data from 2015 to 2020 and discuss its performance.

```python
# visualize original dataset
sns.regplot(data=mex_poultry_df, x='TIME', y='Value') # display the trend line using regplot
plt.show()
```

```python
X = mex_poultry_df['TIME'].to_numpy() # get the explanatory variable
X = X.reshape(-1, 1)

y = mex_poultry_df['Value'].to_numpy()  # get the response variable
y = y.reshape(-1, 1)

lr_model = LinearRegression(fit_intercept=True) # initialize the regression model
lr_model = lr_model.fit(X, y)                   # fit the data to the model i.e. find the line of best fit

# display the parameters for the slope and intercept
print('The intercept is: ', lr_model.intercept_[0])
print('The slope is: ', lr_model.coef_[0,0])

# make a prediction
future_years = np.array(range(2015, 2021)).reshape(-1, 1)
predictions = lr_model.predict(future_years)

# get the actual data from 2015 - 2020
actual_data = df[(df['LOCATION'] == 'MEX') & (df['SUBJECT'] == 'POULTRY') & (df['TIME'].between(2015, 2020)) & (df['MEASURE'] == 'THND_TONNE')][['TIME', 'Value']]
actual_data['Value'] = actual_data['Value'].round(2)
```

```python
# visualize original dataset alongside model's predictions
plt.plot(mex_poultry_df['TIME'], mex_poultry_df['Value']) # plot data for original years
plt.plot(future_years, predictions, label='Predicted Values (2015-2020)') # plot predictions
plt.plot(actual_data['TIME'], actual_data['Value'], label='Actual Values (2015-2020)', color='green')

plt.xlabel('Year') # label x axis
plt.ylabel('Poultry Consumption (thousand tonnes)') # label y axis
plt.title("Mexico's Poultry Consumption Data") # label chart
plt.legend()
plt.grid(True) # show grid for visual boost
plt.show() # show chart
```

The graph above shows Mexico's Poultry Consumption Data from 1994 to 2020. We used a simple linear regression model to predict the values from 2015-2020 and compare them to their actual values. Based on the graph above, we found that the predicted values were higher than the actual values and overestimates Mexico's poultry consumption growth. However, it was still a pretty accurate prediction, as the predicted and actual values both have an upward trend. The difference in values can come from many factors, such as economic conditions or even global conditions. 

## Question 7: 

(5 pts) Calculate the R-squared and explain what this tells you about the model.


```python
#get r-squared, i.e. the coefficient of determination
r_sqrd = lr_model.score(X, y)
r_sqrd = round(r_sqrd, 2)

print('r-squared is: ', r_sqrd)
```

The value of r-squared is 0.98, and this tells us the model is very reliable for finding trends, since 98% of the data is correct. It also tells us that it can be incorrect, as shown in the previous question. The model cannot account for external road bumps. 

## Question 8: 

(5 pts)[optional/bonus] This is a challenge question that may require some research.
residuals: Visualize the residuals of the regression model in question 6 and explain the chart.

Note: A residual is the difference between the actual and predicted values.

```python
residuals = actual_data['Value'].values - predictions.flatten() # get the residuals by substracting actual by predictions

plt.scatter(actual_data['TIME'], residuals) # plot the data

plt.title("Residuals for Mexico's Poultry Consumption Data (2015-2020)")
plt.xlabel('Year')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid(True, linestyle='--')
plt.show()
```
